# **1. Import Library**

Pada tahap ini, Anda perlu mengimpor beberapa pustaka (library) Python yang dibutuhkan untuk analisis data dan pembangunan model machine learning.

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

# **2. Memuat Dataset dari Hasil Clustering**

Memuat dataset hasil clustering dari file CSV ke dalam variabel DataFrame.

```python
import pandas as pd
df = pd.read_csv("hasil_clustering.csv")
df.head()
```

# **3. Data Splitting**

Tahap Data Splitting bertujuan untuk memisahkan dataset menjadi dua bagian: data latih (training set) dan data uji (test set).

```python
X = df.drop(['Cluster'], axis=1)
y = df[['Cluster']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)
```

```python
X_test
```

# **4. Membangun Model Klasifikasi**


## **a. Membangun Model Klasifikasi**

Setelah memilih algoritma klasifikasi yang sesuai, langkah selanjutnya adalah melatih model menggunakan data latih.

Berikut adalah rekomendasi tahapannya.
1. Pilih algoritma klasifikasi yang sesuai, seperti Logistic Regression, Decision Tree, Random Forest, atau K-Nearest Neighbors (KNN).
2. Latih model menggunakan data latih.

```python
# Simpan model dalam dictionary untuk mempermudah pengelolaan
models = {
    "Decision Tree": DecisionTreeClassifier(criterion="entropy"),  # Model Decision Tree
    "Random Forest": RandomForestClassifier(random_state=42)       # Model Random Forest
}
```

```python
#Simpan hasil prediksi dari setiap model
predictions = {}
```

```python
#Latih setiap model dan lakukan prediksi
for name, model in models.items():
    model.fit(X_train, y_train)  # Melatih model menggunakan data latih
    predictions[name] = model.predict(X_test)  # Simpan hasil prediksi
```

Tulis narasi atau penjelasan algoritma yang Anda gunakan.

Dalam membangun model ini, saya menggunakan dua algoritma populer dari Machine Learning untuk klasifikasi, yaitu Decision Tree dan Random Forest.

## **b. Evaluasi Model Klasifikasi**

Berikut adalah **rekomendasi** tahapannya.
1. Lakukan prediksi menggunakan data uji.
2. Hitung metrik evaluasi seperti Accuracy dan F1-Score (Opsional: Precision dan Recall).
3. Buat confusion matrix untuk melihat detail prediksi benar dan salah.

```python
#Evaluasi performa setiap model
for name, y_pred in predictions.items():
    print(f"\nEvaluasi Model: {name}")  # Tampilkan nama model
    print("Confusion Matrix:")
    print(confusion_matrix(y_test, y_pred))  # Tampilkan Confusion Matrix
    print("Classification Report:")
    print(classification_report(y_test, y_pred))  # Tampilkan Classification Report
```

Tulis hasil evaluasi algoritma yang digunakan, jika Anda menggunakan 2 algoritma, maka bandingkan hasilnya.

- Random Forest menunjukkan performa yang sedikit lebih baik dibandingkan Decision Tree terutama pada F1-score dan pengurangan kesalahan prediksi.

- Random Forest lebih unggul karena menggunakan ensemble (gabungan banyak pohon), sehingga mampu memberikan hasil yang lebih stabil dan akurat dibandingkan Decision Tree yang hanya bergantung pada satu pohon keputusan.

- Namun, jika waktu pelatihan dan prediksi menjadi kendala, Decision Tree dapat menjadi pilihan yang lebih cepat dan sederhana.
